```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Classification.csv")
df
```

```python
x = df.iloc[:,:-1]
y = df["Exited"]
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 19)
```

```python
# Initialize model
ann = Sequential()

#Add hidden layer
ann.add(Dense(units = 6, activation = "relu"))

#Add output layer
ann.add(Dense(units = 1, activation = "sigmoid"))

#Establish connection
ann.compile(optimizer = "adam", loss = "binary_crossentropy")

#Train the model
ann.fit(x_train, y_train, epochs = 50 )
```

```python
ypred = ann.predict(x_test)
ypred = ypred>0.5
ypred
```

```python
print(classification_report(y_test, ypred))
```

```python
re = pd.read_csv("https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Regression.csv")
re
```

```python
X = re.iloc[:, 1:]
Y = re["price"]
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y, test_size = 0.22, random_state = 10)
```

```python
rnn = Sequential()
rnn.add(Dense(units = 30, activation = "relu"))
rnn.add(Dense(units = 19, activation = "relu"))
rnn.add(Dense(units = 1))
rnn.compile(optimizer = "adam", loss = "mse")
rnn.fit(Xtrain, Ytrain, epochs = 105)
```

```python
y_pred = rnn.predict(Xtest)
y_pred
```

```python
print(r2_score(Ytest, y_pred))
```
